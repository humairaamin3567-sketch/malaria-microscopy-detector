# 01 - Exploration


This notebook contains exploratory data analysis steps: show sample images, class distribution, and augmentations.

```python
# Example: display a few images (requires data)
from pathlib import Path
from IPython.display import display
from PIL import Image
p=Path('data/raw')
for cls in ['infected','uninfected']:
    d=p/cls
    if d.exists():
        imgs=list(d.glob('*.png'))[:3]
        for i in imgs:
            display(Image.open(i))
```
